```python
import pandas as pd
```

```python
# Drop rows where 'Quantity' and Unit Price is negative
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print(df)
df.shape
```

```python
# Extract the first 5 digits from the StockCode column and overwrite the original column
df['StockCode'] = df['StockCode'].str[:5]
print(df)
df.shape
```

```python
# Check for missing values in the 'StockCode' column and remove rows with NaN values
df = df.dropna(subset=['StockCode'])

# Filter out rows where StockCode is entirely a string
df = df[df['StockCode'].str.isnumeric()]

df.reset_index(drop=True, inplace=True)
df.shape
```

```python
#2) Create a Dictionary with the 10 countries that have the most 'Quantity' ordered except UK

df_exceptUK = df[df['Country'] != 'United Kingdom']

quant = df_exceptUK.groupby('Country')['Quantity'].sum()

# Get the top 10 countries with the most 'Quantity' ordered
top_10_countries = quant.nlargest(10)

# Convert the top 10 countries Series to a dictionary
top_10_countries_dict = top_10_countries.to_dict()

# Display the dictionary
print(top_10_countries_dict)
```

```python
# 3) Find Total Sale between two dates entered

import datetime
# Convert 'InvoiceDate' to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%y %H:%M')

# Input start and end dates from the user
start_date_input = input("Enter the start date (MM/DD/YY): ")
end_date_input = input("Enter the end date (MM/DD/YY): ")

# Convert user input to datetime format
start_date = pd.to_datetime(start_date_input, format='%m/%d/%y')
end_date = pd.to_datetime(end_date_input, format='%m/%d/%y')

# Filter the DataFrame for the user-defined date range
filtered_df = df[(df['InvoiceDate'] >= start_date) & (df['InvoiceDate'] <= end_date)]

# Calculate the total sale by summing the product of 'Quantity' and 'UnitPrice'
total_sale = (filtered_df['Quantity'] * filtered_df['UnitPrice']).sum()

print(f"Total Sale between {start_date.date()} and {end_date.date()}: ${total_sale:.2f}")
```

```python
#4) Find each Countries Total SALE to find which Countries is highest potential customer

# Calculating total sale
df['TotalSale'] = df['Quantity'] * df['UnitPrice']

# Group by 'Country' and calculating the total sale for each country
country_total_sales = df.groupby('Country')['TotalSale'].sum().reset_index()

# Finding the country with the highest total sale
highest_potential_customer = country_total_sales.loc[country_total_sales['TotalSale'].idxmax()]

# Display the total sales for each country and the highest potential customer
print("Total Sales by Country:")
print(country_total_sales.sort_values(by=['TotalSale'], ascending=False).head())
print("\nCountry with the Highest Total Sale:")
print(highest_potential_customer)
```

```python
# 5) Which Customer Id has highest purchase in quantity and total and find out the country of it

# Groupby 'CustomerID' and calculate the total quantity for each customer
customer_highest_quantity = df.groupby('CustomerID')['Quantity'].sum().reset_index()

# Find the customer with the highest total sale
highest_purchase_customerID = customer_highest_quantity.loc[customer_highest_quantity['Quantity'].idxmax()]

# Find the country of the customer with the highest purchase in quantity
country_of_highest_purchase_customer = df[df['CustomerID'] == highest_purchase_customerID['CustomerID']]['Country'].iloc[0]

# Display the total sales for each country and the highest potential customer
print("Quantity Total by Customer:")
print(customer_highest_quantity.sort_values(by=['Quantity'], ascending=False).head())
print("\nCustomer with the Highest Purchase:")
print(highest_purchase_customerID)
print('\nCustomer Country:', country_of_highest_purchase_customer)

```

```python
#6) Which Customer Id has lowest purchase in quantity and total and find out the country of it

# Group by CustomerID and calculate the sum of Quantity for each customer
customer_quantity = df.groupby('CustomerID')['Quantity'].sum()

# Find the Customer ID with the lowest purchase in quantity
min_quantity_customer_id = customer_quantity.idxmin()

# Find the country of the customer with the lowest purchase in quantity
country_of_min_quantity_customer = df[df['CustomerID'] == min_quantity_customer_id]['Country'].iloc[0]

print("Quantity Total by Customer:")
print(customer_highest_quantity.sort_values(by=['Quantity'], ascending=False).head())
print("\nCustomer with the Lowest Purchase:")
print(min_quantity_customer_id)
print('\nCustomer Country:')
print(country_of_min_quantity_customer)
```

```python
#7) Find the biggest order In Sale (Cost) which country was the order from and fiind total purchase (Cost) done by that customer ID

# Group by InvoiceNo and calculate the total sale cost for each invoice
invoice_total_sale = df.groupby('InvoiceNo')['TotalSale'].sum()

# Find the invoice with the highest total sale cost
biggest_order_invoice = invoice_total_sale.idxmax()

# Find the country of the biggest order
country_of_biggest_order = df[df['InvoiceNo'] == biggest_order_invoice]['Country'].iloc[0]

# Calculate the total purchase cost done by the customer responsible for the biggest order
customer_id_of_biggest_order = df[df['InvoiceNo'] == biggest_order_invoice]['CustomerID'].iloc[0]
total_purchase_cost_by_customer = df[df['CustomerID'] == customer_id_of_biggest_order]['TotalSale'].sum()

print(f"Biggest order in Sale: {total_purchase_cost_by_customer:.2f}")
print(f"Biggest Order country: {country_of_biggest_order}.")
print(f"Therefore, the total purchase cost done by customer ID {customer_id_of_biggest_order} for the biggest order is {total_purchase_cost_by_customer:.2f}")
```

```python
#8) Find how many different customers are available for each country save the count in dictionary also return countries with highest and lowest customer count

# Group by 'Country' and calculate the count of unique 'CustomerID' values within each group
customer_counts_by_country = df.groupby('Country')['CustomerID'].nunique()

# Convert the result to a dictionary
customer_counts_dict = customer_counts_by_country.to_dict()

# Find countries with the highest and lowest customer counts
country_with_highest_customers = customer_counts_by_country.idxmax()
country_with_lowest_customers = customer_counts_by_country.idxmin()

# Print the dictionary and the countries with the highest and lowest customer counts
print("Customer Counts by Country:")
print(customer_counts_dict)
print()
print(f"Country with the highest customer count: {country_with_highest_customers}")
print(f"Country with the lowest customer count: {country_with_lowest_customers}")
```

```python
#9) Find which stock code was ordered the most by each country

# Group by 'Country' and 'StockCode', and calculate the count of each stock code within each country
most_ordered_stock = df.groupby(['Country', 'StockCode']).size().reset_index(name='OrderCount')

# Find the stock code with the highest order count for each country
most_ordered_stock = most_ordered_stock.sort_values(by=['Country', 'OrderCount'], ascending=[True, False])
most_ordered_stock = most_ordered_stock.groupby('Country').first().reset_index()

most_stocks = most_ordered_stock.sort_values(by='OrderCount', ascending=False)
print(most_stocks)
print()

print('The stock code that was ordered the most is:')
print(most_stocks.iloc[0])
```
