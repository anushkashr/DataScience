```python
# Creating a tic-tac-toe game between computer and human
# 1st user will enter his name
# user will choose X or O
# randomly first turn will be decided
# After each turn show Board with replaced Indexes by X or O
# If no one wins and board is full say its a draw
# Win conditions are horizontal match or vertical match or diagonal match
# ask user if we wants to play again at end of game import random


import random

def display_board(board_list):
    # Display the tic-tac-toe board
    print(" {} | {} | {} ".format(board_list[0], board_list[1], board_list[2]))
    print("-----------")
    print(" {} | {} | {} ".format(board_list[3], board_list[4], board_list[5]))
    print("-----------")
    print(" {} | {} | {} ".format(board_list[6], board_list[7], board_list[8]))

def check_winner(board, player):
    # Check if the given player has won
    win_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]              # Diagonals
    ]

    for combo in win_combinations:
        # If all positions in a winning combination are occupied by the player's symbol, they win
        if all(board[i] == player for i in combo):
            return True
    return False

def is_board_full(board):
    # Check if the board is fully occupied
    return all(cell != ' ' for cell in board)

def get_player_move(board):
    while True:
        move = input("Enter your move (1-9): ")
        print()
        if move.isdigit() and 1 <= int(move) <= 9:
            index = int(move) - 1
            if board[index] == ' ':
                return index
            else:
                print("Invalid move. Position is already full. Try again.")
        else:
            print("Invalid move. Please enter a number between 1-9.")

def play_game():
    print('-----------------------')
    print("Welcome to Tic-Tac-Toe!")
    print('-----------------------')
    print()

    player_name = input("Enter your name: ")

    # Choose player's symbol (X or O)
    while True:
        player_symbol = input("Choose X or O: ").upper()
        if player_symbol in ['X', 'O']:
            break
        else:
            print("Invalid choice. Choose either 'X' or 'O'.")

    # Assign computer's symbol
    if player_symbol == 'O':
        computer_symbol = 'X'
    else:
        computer_symbol = 'O'

    # Randomly choose who goes first
    current_player = random.choice([player_name, 'Computer'])
    print(f"{current_player} goes first!")
    print()

    # Initialize the game board
    board = [' '] * 9
    display_board(board)
    print()

    while True:
        if current_player == 'Computer':
            print("Computer's turn!")
            print()
            empty_indices = []
            for i, cell in enumerate(board):
                if cell == ' ':
                    empty_indices.append(i)
            move = random.choice(empty_indices)
        else:
            # Get the player's move
            move = get_player_move(board)

        # Update the board with the player's/computer's move
        if current_player == player_name:
            board[move] = player_symbol
        else:
            board[move] = computer_symbol

        # Display the updated board
        display_board(board)
        print()

        # Check for a win, lose, or draw
        if check_winner(board, player_symbol):
            print(f"{player_name} wins!")
            print()
            break
        elif check_winner(board, computer_symbol):
            print("Computer wins!")
            print()
            break
        elif is_board_full(board):
            print("It's a draw!")
            print()
            break

        # Switch players for the next turn
        if current_player == player_name:
            current_player = 'Computer'
        else:
            current_player = player_name



        # Ask if the player wants to play again
    while True:
            play_again = input("Do you want to play again? (yes/no): ")
            if play_again.lower() == 'yes':
                play_game()
            elif play_again.lower() == 'no':
                print("Thanks for playing!")
                break
            else:
              print("Invalid input. Please enter 'yes' or 'no'")

play_game()
```
