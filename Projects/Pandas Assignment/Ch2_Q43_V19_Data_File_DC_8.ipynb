```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/content/Ch2_Q43_V19_Data_File.xlsx'
df = pd.read_excel('Ch2_Q43_V19_Data_File.xlsx')
```

```python
df = df.iloc[:,:3]
df
```

```python
# Line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['AMZN'], color='blue', label="Amazon")
plt.plot(df['Date'], df['GOOG'], color='red', label="Google")
plt.xlabel('Company')
plt.ylabel('Stock Prices')
plt.title('Stock Prices: AMZN vs GOOG')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print('Analyzing the line chart, we can say that Amazon stock price seems to have a greater trajectory.')
```
