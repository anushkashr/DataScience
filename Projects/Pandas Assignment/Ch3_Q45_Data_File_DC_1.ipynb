```python
#Mean of Monthly Rent
mean_rent = df['Rent'].mean().round(2)
mean_rent
```

```python
#Standard Deviation of Monthly Rent
std_rent = df['Rent'].std().round(2)
std_rent
```

```python
#Mean of Footage
mean_footage = df['Footage'].mean().round(2)
mean_footage
```

```python
#Standard Deviation of Footage
std_footage = df['Footage'].std().round(2)
std_footage
```

c.Which sample data exhibit greater relative dispersion?

```python
# Coefficient of variation (CV) for each column
cv_rent = (std_rent / mean_rent) * 100
cv_footage = (std_footage / mean_footage) * 100

print("Coefficient of Variation for Rent:", cv_rent)
print("Coefficient of Variation for Footage:", cv_footage)
print('\nTherefore, Footage exhibits greater relative dispersion')
```
