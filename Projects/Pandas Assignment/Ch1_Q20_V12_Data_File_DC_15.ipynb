```python
import pandas as pd
```

```python
path = '/content/Ch1_Q20_V12_Data_File (1).xlsx'
df = pd.read_excel('Ch1_Q20_V12_Data_File (1).xlsx')
```

```python
df = df.iloc[:,:4]
df
```

```python
#a.Â How many observations haveÂ x1Â values greater than 30?
count_x1_greater_than_30 = df[df['x1'] > 30]['Observation'].count()

print("Number of observations with x1 values greater than 30:", count_x1_greater_than_30)
```

```python
#b.Â Sort the data byÂ x1Â in ascending order. What are theÂ x1,Â x2,Â andÂ x3Â values of the first observation after the data are sorted?

df_sorted = df.sort_values(by='x1', ascending=True)

first_observation_values = df_sorted.iloc[0]

print('x1,Â x2,Â andÂ x3Â values of the first observation after the data are sorted:')
print("x1:", first_observation_values['x1'])
print("x2:", first_observation_values['x2'])
print("x3:", first_observation_values['x3'])
```

```python
# c.Â Sort the data byÂ x1Â in descending order. What are theÂ x1,Â x2,Â andÂ x3Â values of the first observation after the data are sorted?

df_sorted = df.sort_values(by='x1', ascending=False)

first_observation_values = df_sorted.iloc[0]

print('x1,Â x2,Â andÂ x3Â values of the first observation after the data are sorted:')
print("x1:", first_observation_values['x1'])
print("x2:", first_observation_values['x2'])
print("x3:", first_observation_values['x3'])
```

```python
# d.Â How many missing values are there inÂ x1,Â x2,Â andÂ x3?

missing_x1 = df['x1'].isna().sum()
missing_x2 = df['x2'].isna().sum()
missing_x3 = df['x3'].isna().sum()

print("Missing values in x1:", missing_x1)
print("Missing values in x2:", missing_x2)
print("Missing values in x3:", missing_x3)
```
