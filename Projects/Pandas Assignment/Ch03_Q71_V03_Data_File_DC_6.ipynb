```python
import pandas as pd
```

```python
path = '/content/Ch03_Q71_V03_Data-File.xlsx'
df = pd.read_excel('Ch03_Q71_V03_Data-File.xlsx')
```

```python
df_cleaned = df.iloc[:, :2]
df_cleaned
```

```python
df_cleaned.shape
```

```python
df2 = df_cleaned.dropna()
df2
```

```python
#a.Calculate the covariance between the variables.
#(Negative value should be indicated by a minus sign.
#Round your intermediate calculations to at least 4 decimal places and final answer to 2 decimal places)

covariance = df2['x'].cov(df2['y'])
print(covariance.round(4))
print("Covariance between x and y:", covariance.round(2))
```

```python
#b. Calculate the correlation coefficient.
# (Round your intermediate calculations to 4 decimal places and final answer to 2 decimal places.)

# Sstandard deviation of x and y
std_x = df2['x'].std()
std_y = df2['y'].std()

# Correlation coefficient
correlation_coefficient = covariance / (std_x * std_y)

correlation_coefficient = round(correlation_coefficient, 4)
print(correlation_coefficient)

print("Correlation Coefficient between x and y:", correlation_coefficient.round(2))
```
