```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/content/Ch3_Q25_Data_File.xlsx'
df = pd.read_excel('Ch3_Q25_Data_File.xlsx')
```

```python
df.shape
```

```python
df = df.iloc[:,:2]
df
```

```python
percentiles = df['PE_Ratio'].quantile([0.25, 0.5, 0.75])

percentiles = percentiles.round(2)

print("25th Percentile:", percentiles[0.25])
print("50th Percentile:", percentiles[0.5])
print("75th Percentile:", percentiles[0.75])
```

```python
plt.figure(figsize=(8, 4))
plt.boxplot(df['PE_Ratio'], vert=False)
plt.title('Box Plot of PE Ratio')
plt.xlabel('PE Ratio')
plt.show()
```

```python
# b-2. Are there any outliers?

# Calculate the Inter Quartile Range
Q1 = df['PE_Ratio'].quantile(0.25)
Q3 = df['PE_Ratio'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['PE_Ratio'] < lower_bound) | (df['PE_Ratio'] > upper_bound)]

print("Outliers:")
print(outliers)
```

```python
# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['PE_Ratio'], bins=10, color='blue', edgecolor='black')
plt.title('Histogram of PE Ratio')
plt.xlabel('PE Ratio')
plt.ylabel('Frequency')

plt.show()

from scipy.stats import skew
skewness = skew(df['PE_Ratio'])
print("Skewness:", skewness)

if abs(skewness) < 0.5:
    print("The distribution is approximately symmetric.")
else:
    print("The distribution is not symmetric.")
```
