```python
class dog_info:
  def __init__(self, id, name, breed, color, dob):
    self.id = id
    self.name = name
    self.breed = breed
    self.color = color
    self.dob = dob

    print('ID:', self.id)
    print('Name:',self.name)
    print('Breed:', self.breed)
    print('Color:', self.color)
    print('Date of Birth:', self.dob)
```

```python
dog1 = dog_info(1, 'Bruno', 'Golden Retriever', 'Brown', 1995)
```

```python
dog2 = dog_info(2, 'Marla', 'Husky', 'Grey', 1993)
```

```python
class Car:
  def __init__(self, name, model, fuel, cc, cylinders):
    self.name = name
    self.model = model
    self.fuel = fuel
    self.cc = cc
    self.cylinders = cylinders

  def car_details(self):
    print('Name:', self.name)
    print('Model:',self.model)
    print('Fuel:', self.fuel)
    print('CC:', self.cc)
    print('Cylinders:', self.cylinders)
```

```python
car1 = Car('BMW', 555, 'Electric', 3500, 12)
```

```python
car1.car_details()
```
