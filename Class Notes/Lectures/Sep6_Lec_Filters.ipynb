```python
#Write a query so that $ should be removed and value should be converted into float value

for value in df_chipotle.item_price:
  print(value[1:-1])

#skipping $ and picking everything except $
prices = [float(value[1:-1]) for value in df_chipotle.item_price]
df_chipotle.item_price = prices

df_chipotle

```
