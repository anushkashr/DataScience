

*   Series is one-dimensional data
*   Dataframe is multi-dimensional data



```python
import pandas as pd

lst1 = [12,23,34,45,56,67,78,89,90,110]
sr1 = pd.Series(lst1)
sr1, sr1.ndim, type(sr1), sr1.size, sr1.dtype
```

```python
tup1 = (12,23,34,45,56,67,78,89,90,110)
sr1 = pd.Series(tup1)
sr1, sr1.ndim, type(sr1), sr1.size, sr1.dtype
```

Series Doesn't work with set

```python
set1 = {'a', 'b', 'c', 'd', 'e', 'f'}
sr1 = pd.Series(set1)
sr1, sr1.ndim, type(sr1), sr1.size, sr1.dtype
```

Ordered: can be re-arranged
Unordered: cannot be

```python
dict1 = {'a': 'apple', 'b':'banana', 'c':'cider', 'd': 'dragonfruit'}
sr1 = pd.Series(dict1)
sr1, sr1.ndim, type(sr1), sr1.size, sr1.dtype
```

```python
lst1 = [12,23,34,45,56,67,78,89,90,110]
sr1 = pd.Series(lst1)
sr1, len(sr1), sr1.count(), sr1.ndim, type(sr1), sr1.size, sr1.dtype, sr1.item, sr1.items, sr1.nbytes, sr1.shape
```
